#4. 딕셔너리

```python
abc = 'aaabbccbbaaa'

abc_count = {}

for char in abc:
    if char in abc_count.keys():
        abc_count[char] = abc_count[char] + 1
    else:
        abc_count[char] = 1

for key, value in abc_count.items():
    print(f'abc have {value} {key}.')
```

```python
print('abc_count :', abc_count)
print('abc_count.keys() :', abc_count.keys())
print('abc_count.values() :', abc_count.values())
print('abc_count.items() :', abc_count.items())
```

#6. Function

```python
def add(a, b):
    return a + b

def HelloWorld():
    print("Hello World!")

def OneTwoThree():
    return 1, 2, 3

print(f'{3} + {4} = {add(3, 4)}')
HelloWorld()

one, two, three = OneTwoThree()
print(f'one : {one}')
print(f'two : {two}')
print(f'three : {three}')
```

```python
# positional argument

def Print(name, age, school):
    print('name:', name)
    print('age:', age)
    print('school:', school)

Print('star._.overflow', 23, 'Korea Univ.')
```

```python
# keyword argument

Print(age=23, name='star._.overflow', school='Korea Univ.')
print()
Print('star._.overflow', school='Korea Univ.', age=23)
```

```python
# positional argument Unpacking
def abc(a, b, c):
    print("a :", a)
    print("b :", b)
    print("c :", c)

p = [2, 5, 9]
abc(*p)
print()
abc(*[4, 2, 7])
```

```python
def abc2(*args):
    for i, arg in enumerate(args):
        print(f'{i}th arg is {arg}.')
    
abc2(1, 2, 3, 4, 5)
print()
abc2(2, 5, 7)
```

```python
# keyword argument Unpacking

args = {'a': 4, 'b': 7, 'c': 2}

abc(**args)
```

```python
def abc3(**args):
    for key, value in args.items():
        print(f'{key} : {value}')
    
abc3(**args)
print()
abc3(**{'a': 5, 'b': 3, 'c': 9})
```

```python
# lambda

L = lambda x : x ** 2

x = [1, 2, 3, 4, 5]

xx = [L(x_) for x_ in x]

for k in zip(x, xx):
    print('{}^2 = {}'.format(k[0], k[1]))
```

# 7. Generator

```python
def Range(start, end, step=1):
    i = start

    while i < end:
        yield i
        i = i + step

for i, num in enumerate(Range(start=0, end=10)):
    print(f'{i}th num is {num}')
```

```python
import collections.abc as abc

gen = Range(0, 10, 2)

if isinstance(gen, abc.Iterable):
    print('gen is Iterable.')

for i in range(10):
    print(next(gen))
```

```python
def gen_from(list_num):
    yield from list_num

list_num = [2, 5, 4, 7, 9]

for i, num in enumerate(gen_from(list_num)):
    print(f'{i}th num is {num}.')
```

```python
class Basic:
    def __init__(self):
        self._name = 'Basic'
    
    def name(self):
        return self._name

basic = Basic()
print("Name of basic :", basic.name())
```

```python
# class attribute

class SharedList:
    shared_list = []

    def append(self, num):
        SharedList.shared_list.append(num)
    
    def print(self):
        print(SharedList.shared_list)

a = SharedList()
b = SharedList()

a.append(1)
b.append(2)

a.print()

a.append(3)
b.append(4)

b.print()
```

```python
# inheritance

class Nation:
    name = 'name of nation'
    population = 'population of nation'
    capital = 'capital of nation'

    def show(self):
        print('Method of Nation class')
    
class Korea(Nation):
    def __init__(self, name, population, capital):
        self.name = name
        self.population = population
        self.capital = capital
    
    def show_name(self):
        print('name of nation :', self.name)
        print('population of nation :', self.population)
        print('capital of nation :', self.capital)

    def show(self):
        print("======================")
        super().show()
        print("======================")

        print(
            '''
Method Overriding

name of nation : {}
population of nation : {}
capital of nation : {}
            '''.format(self.name, self.population, self.capital)
        )
    
korea = Korea('대한민국', '5천만 명', '서울')
korea.show()
print()
korea.show_name()
```

```python
# static method

class Add:
    @staticmethod
    def add(a, b):
        return a + b
    
print('{} + {} = {}'.format(3, 4, Add.add(3, 4)))
```

```python
# class method

class Sum:
    total_sum = 0
    @classmethod
    def sum(cls, num):
        cls.total_sum = cls.total_sum + num
    
    @classmethod
    def total(cls):
        return cls.total_sum
    
a = Sum()
b = Sum()

a.sum(3)
b.sum(5)

print(a.total())

a.sum(7)
b.sum(9)

print(b.total())
```

# 9. Exception

```python
def str2int(s):
    try:
        i = int(s)
        print('str2int success')
    except ValueError:
        i = -1
        print('str2int fail')
    except TypeError:
        i = -1
        print('str2int fail')
    return i

print(str2int('123'))
print()
print(str2int('asdf'))
print()
print(str2int([1, 2, 3]))
```

# 10. with

```python
class Program:
    def __init__(self):
        self.check = False

    def open(self):
        self.check = True
        print("Program start")

    def print(self, content):
        if self.check:
            print(content)
        else:
            print("Can't print content")

    def close(self):
        self.check = False
        print("Program end")

program = Program()

try:
    program.open()
    program.print('asdf')
    program.print('sdfg')
    program.print('dfgh')
finally:
    program.close()

program.print('asdf')
```

```python
class Program2(Program):
    def __init__(self):
        super().__init__()
    
    def __enter__(self):
        self.check = True
        print("Program2.__enter__()")
        return self
    
    def print(self, content):
        if self.check:
            print(content)
        else:
            print("Can't print content")
    
    def __exit__(self, type, value, traceback):
        self.check = False
        print("Program2.__exit__()")

with Program2() as program2:
    program2.print("asdf")
    program2.print("fdsa")

program2.print("asdf")
```
